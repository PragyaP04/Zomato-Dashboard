Using numpy library we convert invalid values('New' and '-') into missing values (NaN)

```python
df['rate'] = df['rate'].replace('NEW', np.nan)
df['rate'] = df['rate'].replace('-', np.nan)
```

Split the rating string at '/' and keep only the first part (actual rating value)

```python
df['rate'] = df['rate'].str.split('/').str[0]
```

Rate is string data type and we are converting it into a float datatype to perform rest of the operations

```python
df['rate'] = df['rate'].astype(float)
```

```python
df['rate'].head()
```

```python
df['rate'].dtype
```

This part evaluates how many missing values are present in the table

```python
df.isnull().sum()
```

Difference between drop and dropna


*   in drop we drop the whole column
*   in dropna we just drop the rows where the data is missing




*   We are droping the columns which are not usefull in the data.
*   In pandas axis=0 means rows and axis=1 means colums so we are doing here axis1 so that we can remove the entire column that we do not need further in our analysis


*   We are using here inplace means we do not want to create a different dataframe with the columns dropped we want the changes in our current dataframe






```python
df.drop(['url', 'address', 'phone', 'menu_item', 'reviews_list'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df = df.dropna(subset=['rate', 'location', 'cuisines'])
```

In fillna we dont replace delete rows we just replace missing dish names with Not Mentioned

```python
df['dish_liked'] = df['dish_liked'].fillna('Not Mentioned')
```

```python
df.isnull().sum()
```

here too we fill the unknown missing values too unknown

```python
df['rest_type'] = df['rest_type'].fillna('Unknown')
```

```python
df = df.dropna(subset=['approx_cost(for two people)'])
```

```python
df.isnull().sum()
```
